--------------------

import json --> תייבא קובץ גייסון (ששמרנו על המחשב שלנו)
--------------------

import openpyxl --> תייבא קובץ אקסל (ששמרנו על המחשב שלנו)
--------------------

from openpyxl.styles import Font, PatternFill, Alignment, Border, Side ----> מספריית העיצוב - תייבא פונקציות של פונט, מילוי תא, מרכוז תא, גבול תא, צד (עבור הדגשת תא)

פונט - נגדיר אם הפונט בולד, ואת צבע הפונט

מילוי תא - נגדיר את צבע הרקע של התא והאם כולו יהיה באותו צבע (סוליד)

מרכוז תא - נגדיר את מיקום הטקסט בתא (בצד/במרכז)

גבול תא - נגדיר את עובי הגבול של התא בכל כיוון (בחלק עליון/חלק תחתון/בצד ימין/צד שמאל)

--------------------
from openpyxl.chart import BarChart, LineChart, Reference --> מספריית הגרפים - תייבא פונקציות

BarChart -  (עמודות)נגדיר את סוג הגרף

LineChart - גרף קו

Reference - מאיפה הוא לוקח את הדאטה

```python
'''
קובץ אנליזות בית קפה

'''
import json 
import openpyxl 
from openpyxl.styles import Font,PatternFill,Alignment,Border,Side#ספריות פונט, מילוי תא, מרכוז, גבול
from openpyxl.chart import BarChart,LineChart,Reference#רפרנס-נתונים שלתוכם אני צריך לשים כדי לקחת לגרף

def jsonToExcelMultiSheet(json_file,excel_file):   #הפונקציה מקבלת json file-שם הקובץ של ה-json, קובץ אקסל-שם האקסל שאצור
       #step 1:Read JSON File
       print(f'Reading data from {json_file}')
       with open(json_file,'r',encoding='utf-8') as f: #encoding-אם יש עברית, המחשב ידע לקדד ולהציג עברית
          data=json.load(f)#טעינת הקובץ
        
        #step 2:Create Excel Workbook
       print(f'creating Excel Workbook with multiple sheets...')
       wb=openpyxl.Workbook()#יצירת חוברת עבודה חדשה באקסל
       wb.remove(wb.active)#Reomve default sheet

        #Define border style for all sheets
       thin_border=Border(
            left=Side(style='thin'),#צד שמאל של התא
            right=Side(style='thin'),#צד ימין של התא
            top=Side(style='thin'),#צד עליון של התא
            bottom=Side(style='thin')#צד תחתון של התא
        )
        #============================
        #Sheet 1:overview
        #============================
       print(f'Creating Overview Sheet')
       ws_overview=wb.create_sheet("overview")#יצירת גיליון חדש עם שם ספציפי-כל פעם שאפנה אדע לאיזה התכוון
        #Title
       ws_overview.merge_cells('A1:D1')#מאחד את התאים מ- A1 עד D1-merge.cells=איחוד תאים-הפיכת התא לתא אחד
       ws_overview['A1']=f"{data['shop_info']['name']}- Business Overview"#מציב בתא את שם החנות כפי שמופיע בקובץ של ה-json
       ws_overview['A1'].font=Font(size=16,bold=True,color='FFFFFF')#עיצוב הפונט עם גודל 16, טקסט מודגש, צבע לבן-עפ"י המטלה
       ws_overview['A1'].fill=PatternFill(start_color='1F4E78',fill_type='solid')#עיצוב המילוי עם רקע בצבע כחול, solid-כל התא- אפשר גם end color
       ws_overview['A1'].alignment=Alignment(horizontal='center')#ממקם את הטקסט בתא שיהיה במרכז- אפשר גם vertical-left,right,center-horizental or top center bottom-vertical

       #Shop Info
       ws_overview['A3']="Location"#כותרת מיקום
       ws_overview['B3']=data['shop_info']['location']#ניגש לערך של המיקום ושם עמודה מימין באותה שורה
       ws_overview['A4']='Manager'#כותרת מנהל
       ws_overview['B4']=data['shop_info']['manager']#ניגש לערך של המנהל ושם עמודה מימין באותה שורה
       ws_overview['A5']='Established'#כותרת תאריך הקמה
       ws_overview['B5']=data['shop_info']['established']#ניגש לערך של התאריך הקמה ושם עמודה מימין באותה שורה

       for row in range(3,6):#לרוץ על הטווח שהגדרנו-שורות 3 עד 6 לא כולל
          ws_overview[f'A{row}'].font=Font(bold=True)#בעמודה A- הדגשת התוויות-הכותרות

        #column widths
       ws_overview.column_dimensions['A'].width=15#הגדרת רוחב עמודה 1 לרוחב 15
       ws_overview.column_dimensions['B'].width=25# הגדרת רוחב עמודה 2 לרוחב 25
      
        #============================
        #Sheet 2:Monthly sales(with line chart)-גרף קו
        #============================
       print("creating monthly sales sheet with chart")#חיווי בשבילי שהקוד הצליח להגיע עד לפה
       ws_sales=wb.create_sheet("Monthly Sales")#יצירת באותו הקובץ אקסל גיליון חדש
       
       #Title
       #הכותרת היא העתקה למה שעשינו מקודם
       ws_sales.merge_cells('A1:E1')#מאחד את התאים מ- A1 עד E1-merge.cells=איחוד תאים-הפיכת התא לתא אחד
       ws_sales['A1']="Monthly Sales Performance"#כותרת בשורה 1
       ws_sales['A1'].font=Font(size=14,bold=True,color='FFFFFF')#עיצוב הפונט עם גודל 16, טקסט מודגש, צבע לבן-עפ"י המטלה
       ws_sales['A1'].fill=PatternFill(start_color='1F4E78',fill_type='solid')#עיצוב המילוי עם רקע בצבע כחול, solid-כל התא- אפשר גם end color
       ws_sales['A1'].alignment=Alignment(horizontal='center')#ממקם את הטקסט בתא שיהיה במרכז- אפשר גם vertical-left,right,center-horizental or top center bottom-vertical
       #Headers
       #כותרות העמודות-בכל פעם מעתיק ומדביק את כותרות העמודות ומשנה את הערכים בהתאם למה שמבקשים-לשים לב לאותיות קטנות וגדולות
       headers=['Month','Revenue(nis)','Expenses(nis)','Profit(nis)','Customers']#יצירת רשימה לכותרות העמודות
       #מעבר בלולאה על כל עמודה וכותרת ומציב בכל פעם בתא בשורה 3 את ערך האיבר ברשימת הכותרות ומעצב אותו עפ"י הדרישות
       for col,header in enumerate(headers,start=1):#עבור כל איבר ברשימה, עובר על העמודות מהמספור שמתחיל ב-1
             cell=ws_sales.cell(row=3,column=col)#בין הכותרות לטבלה יש שורה ריקה, בכל פעם מציב את הערך לעמודה
             cell.value=header#התוכן של התא יהיה הערך של כל איבר ברשימת הכותרות
             cell.font=Font(bold=True,color='FFFFFF')#פונט של הכתב בצבע לבן
             cell.fill=PatternFill(start_color="4472C4",fill_type='solid')#גוונים-תכלת
             cell.alignment=Alignment(horizontal='center')
             cell.border=thin_border
       #Data-
       #מגדיר שורה 4, שורה לאחר הכותרות
       #עובר בלולאה באינדקס שאני מגדיר על כל הנתונים של המפתח מכירות חודשיות-רשימת המכירות
       #בתחילת הלולאה מחשב לכל חודש את הרווח עפ"י הכנסות והוצאות(משתנה משאלה לשאלה החישוב)
       row=4
       for month_data in data['monthly_sales']:#מעבר על כל מילון שהוא נתונים על חודש-איבר בתוך רשימת המכירות החודשיות
           profit=month_data['revenue']-month_data['expenses']#מחשב רווח שהוא הפרש של ההכנסות פחות ההוצאות-גישה לערך במפתח בכל מילון שהוא איבר ברשימה
           #גישה לכל ערך במפתח במילון של החודש-הצבה בכל פעם בעמודה אחרת באותה שורה את ערך המפתח, עושה פורמט לשקל שיציג בעיצוב לתא שצריך
           ws_sales.cell(row=row,column=1).value=month_data['month']
           ws_sales.cell(row=row,column=2).value=month_data['revenue']
           ws_sales.cell(row=row,column=2).number_format='₪#,##0'
           ws_sales.cell(row=row,column=3).value=month_data['expenses']
           ws_sales.cell(row=row,column=3).number_format='₪#,##0'
           ws_sales.cell(row=row,column=4).value=profit#הצבה לערך שחישבנו, לא היה מוגדר כמפתח במילון בהתחלה
           ws_sales.cell(row=row,column=4).number_format='₪#,##0'
           ws_sales.cell(row=row,column=5).value=month_data['customers']

       #Add Borders
           for col in range(1,6):
              ws_sales.cell(row, column=col).border=thin_border#הוספת מסגרות לכל תאי העמודות שהוספתי בשורה

           row+=1#קידום בכל פעם  בשורה לקראת איטרציה של יצירת שורה נוספת של ערכי המפתחות
          
      #column widths
       for col in ['A','B','C','D','E']:#יצירה לכל עמודה רוחב של 15
         ws_sales.column_dimensions[col].width=15

      #Create Line Chart-יצירת גרף עפ"י בקשות השאלה ב-Monthly_Sales
       chart1=LineChart()
       chart1.title='Monthly Financial Performance'
       chart1.style=10
       chart1.y_axis.title='Amount(nis)'#כותרת לציר ה-y
       chart1.x_axis.title='Month'#כותרת לציר ה-x
       #Data for chart
       data_ref=Reference(ws_sales,min_col=2,min_row=3,max_row=row-1,max_col=4)
       categories=Reference(ws_sales,min_col=1,min_row=4,max_row=row-1)
       
       chart1.add_data(data_ref,titles_from_data=True)
       chart1.set_categories(categories)
       chart1.height=10
       chart1.width=20

       ws_sales.add_chart(chart1,"A12")

       #============================
      #Sheet 3:Products
      #============================

       print("creating products sheet")
       ws_prod=wb.create_sheet("Products")

       #Title
       ws_prod.merge_cells('A1:F1')
       ws_prod['A1']="Monthly Sales Performance"#כותרת בשורה 1
       ws_prod['A1'].font=Font(size=14,bold=True,color='FFFFFF')#עיצוב הפונט עם גודל 16, טקסט מודגש, צבע לבן-עפ"י המטלה
       ws_prod['A1'].fill=PatternFill(start_color='1F4E78',fill_type='solid')#עיצוב המילוי עם רקע בצבע כחול, solid-כל התא- אפשר גם end color
       ws_prod['A1'].alignment=Alignment(horizontal='center')
       #headers
       headers=['Product','category','Price(nis)','Cost(nis)','Units Sold','Total Profit']#יצירת רשימה לכותרות העמודות
       for col,header in enumerate(headers,start=1):#עבור כל איבר ברשימה, עובר על העמודות מהמספור שמתחיל ב-1
             cell=ws_prod.cell(row=3,column=col)#בין הכותרות לטבלה יש שורה ריקה, בכל פעם מציב את הערך לעמודה
             cell.value=header#התוכן של התא יהיה הערך של כל איבר ברשימת הכותרות
             cell.font=Font(bold=True,color='FFFFFF')#פונט של הכתב בצבע לבן
             cell.fill=PatternFill(start_color="70AD47",fill_type='solid')#גוונים-גוון ירוק-ביקשו צבע שונה לכל גיליון-מילוי שונה
             cell.alignment=Alignment(horizontal='center')
             cell.border=thin_border
       row=4#הגדרת שורת התחלה לנתונים לאחר כותרות העמודות
       for product in data['products']:#מעבר על כל מילון שהוא נתונים על חודש-איבר בתוך רשימת המכירות החודשיות
           profit_per_unit=product['price']-product['cost']
           total_profit=profit_per_unit*product['units_sold']#מבצע השמה למשתנה של הרווח הכולל
           #גישה לכל ערך במפתח במילון של החודש
           ws_prod.cell(row=row,column=1).value=product['name']
           ws_prod.cell(row=row,column=2).value=product['category']
           ws_prod.cell(row=row,column=3).value=product['price']
           ws_prod.cell(row=row,column=3).number_format='₪#,##0'
           ws_prod.cell(row=row,column=4).value=product['cost']
           ws_prod.cell(row=row,column=4).number_format='₪#,##0'
           ws_prod.cell(row=row,column=5).value=product['units_sold']
           ws_prod.cell(row=row,column=6).value=total_profit
           ws_prod.cell(row=row,column=6).number_format='₪#,##0'

       #Add Borders
           for col in range(1,7):#עד 6 עמודות לא כולל
              ws_prod.cell(row, column=col).border=thin_border
           row+=1#קידום השורה בכל פעם
       ws_prod.column_dimensions['A'].width=15
       ws_prod.column_dimensions['B'].width=15
       ws_prod.column_dimensions['C'].width=12
       ws_prod.column_dimensions['D'].width=12
       ws_prod.column_dimensions['E'].width=12
       ws_prod.column_dimensions['F'].width=15

       #============================
      #Sheet 4:Employees
      #============================
       print("creating Employees sheet")
       ws_emp=wb.create_sheet('Employees')
       ws_emp.merge_cells('A1:E1')#מיזוג התאים בכל פעם כמספר כותרות העמודות שמבקשים
       ws_emp['A1']="Employee Information"#כותרת בשורה 1
       ws_emp['A1'].font=Font(size=14,bold=True,color='FFFFFF')#עיצוב הפונט עם גודל 16, טקסט מודגש, צבע לבן-עפ"י המטלה
       ws_emp['A1'].fill=PatternFill(start_color='1F4E78',fill_type='solid')#עיצוב המילוי עם רקע בצבע כחול, solid-כל התא- אפשר גם end color
       ws_emp['A1'].alignment=Alignment(horizontal='center')
       #headers
       headers=['Name','Position','Hours/Week','Hourly Rate(nis)','Weekly Salary(nis)']#יצירת רשימה לכותרות העמודות
       for col,header in enumerate(headers,start=1):#עבור כל איבר ברשימה, עובר על העמודות מהמספור שמתחיל ב-1
             cell=ws_emp.cell(row=3,column=col)#בין הכותרות לטבלה יש שורה ריקה, בכל פעם מציב את הערך לעמודה
             cell.value=header#התוכן של התא יהיה הערך של כל איבר ברשימת הכותרות
             cell.font=Font(bold=True,color='FFFFFF')#פונט של הכתב בצבע לבן
             cell.fill=PatternFill(start_color="FFC000",fill_type='solid')#גוונים-צבע כתום
             cell.alignment=Alignment(horizontal='center')
             cell.border=thin_border
       #Data
       row=4
       total_weekly_cost=0
       for employee in data['employees']:#מעבר על כל מילון שהוא נתונים על חודש-איבר בתוך רשימת המכירות החודשיות
           weekly_salary=employee['hours_per_week']*employee['hourly_rate']
           total_weekly_cost+=weekly_salary#מחשבים עבור כל עובד בנפרד את המשכורת השבועית וצוברים את זה בסה"כ המשכורת השבועית
           #גישה לכל ערך במפתח במילון של החודש
           ws_emp.cell(row=row,column=1).value=employee['name']
           ws_emp.cell(row=row,column=2).value=employee['position']
           ws_emp.cell(row=row,column=3).value=employee['hours_per_week']
           ws_emp.cell(row=row,column=4).value=employee['hourly_rate']
           ws_emp.cell(row=row,column=4).number_format='₪#,##0'
           ws_emp.cell(row=row,column=5).value=weekly_salary
           ws_emp.cell(row=row,column=5).number_format='₪#,##0'
           if ws_emp.cell(row=row,column=5).value>2000:
              ws_emp.cell(row=row,column=5).fill=PatternFill(start_color='70AD47',fill_type='solid')
           elif ws_emp.cell(row=row,column=5).value>1600 and ws_emp.cell(row=row,column=5).value<2000:
              ws_emp.cell(row=row,column=5).fill=PatternFill(start_color='F3A1C7',fill_type='solid')
           else:
              ws_emp.cell(row=row,column=5).fill=PatternFill(start_color='FFC000',fill_type='solid')




       #Add Borders
           for col in range(1,6):
              ws_prod.cell(row, column=col).border=thin_border
           row+=1
        #Total Row-הוספת שורת הסיכום בתחתית המציגה את העלות השבועית הכוללת
       ws_emp.cell(row=row,column=4).value="Total Weekly Cost"
       ws_emp.cell(row=row,column=4).font=Font(bold=True)
       ws_emp.cell(row=row,column=5).value=total_weekly_cost
       ws_emp.cell(row=row,column=5).number_format='₪#,##0'
       ws_emp.cell(row=row,column=5).font=Font(bold=True)
       ws_emp.cell(row=row,column=5).fill=PatternFill(start_color="FFC000",fill_type='solid')

       for col in ['A','B','C','D','E']:#יצירה לכל עמודה רוחב של 25
          ws_emp.column_dimensions[col].width=25
       

       #============================
      #Sheet 5:Customer Ratings
      #============================
      #יצירת גליון
       print("creating customers Ratings sheet")
       ws_cust=wb.create_sheet("Customer Ratings")
       #Title
       ws_cust.merge_cells('A1:D1')
       ws_cust['A1']="Customers Satisfaction Ratings"#כותרת בשורה 1
       ws_cust['A1'].font=Font(size=14,bold=True,color='FFFFFF')#עיצוב הפונט עם גודל 16, טקסט מודגש, צבע לבן-עפ"י המטלה
       ws_cust['A1'].fill=PatternFill(start_color='1F4E78',fill_type='solid')#עיצוב המילוי עם רקע בצבע כחול, solid-כל התא- אפשר גם end color
       ws_cust['A1'].alignment=Alignment(horizontal='center')
      #headers
       headers=['Category','Rating(out of 5)','Number of Reviews']#יצירת רשימה לכותרות העמודות
       for col,header in enumerate(headers,start=1):#עבור כל איבר ברשימה, עובר על העמודות מהמספור שמתחיל ב-1
             cell=ws_cust.cell(row=3,column=col)#בין הכותרות לטבלה יש שורה ריקה, בכל פעם מציב את הערך לעמודה
             cell.value=header#התוכן של התא יהיה הערך של כל איבר ברשימת הכותרות
             cell.font=Font(bold=True,color='FFFFFF')#פונט של הכתב בצבע לבן
             cell.fill=PatternFill(start_color="A2DBFF",fill_type='solid')#גוונים-צבע כתום
             cell.alignment=Alignment(horizontal='center')
             cell.border=thin_border
      #Data
       row=4
       for customer in data['customer_ratings']:#מעבר על כל מילון שהוא נתונים על חודש-איבר בתוך רשימת המכירות החודשיות
           #גישה לכל ערך במפתח במילון של החודש
           ws_cust.cell(row=row,column=1).value=customer['category']
           ws_cust.cell(row=row,column=2).value=customer['rating']
           ws_cust.cell(row=row,column=3).value=customer['reviews']
           #Add Borders
           for col in range(1,4):#
              ws_cust.cell(row, column=col).border=thin_border
           row+=1 
       for col in ['A','B','C']:
             ws_cust.column_dimensions[col].width=25
    
 

 

       wb.save(excel_file)
       





jsonToExcelMultiSheet('coffee_shop_data.json','coffee_shop_data.xlsx')#מזמן את הפונקציה עם שם הפונקציה ושם הקובץ אקסל

```
